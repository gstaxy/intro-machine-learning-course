### Importer les librairies nécessaires

```python
# Importer les librairies de base
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Définir le style des graphiques
sns.set_style("darkgrid")

# Importer les librairies d'apprentissage automatiquefrom sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
```

### Importer le jeu de données Diabiètes pour la classification

```python
# Importer la librairie pandas
import pandas as pd

# Télécharger le jeu de données
data = pd.read_csv('diabetes.csv')

# Remplacer le nom des colonnes
colnames = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
data.columns = colnames

data.head()
```

### Séparer les données en sous-ensemble X et y

```python
# Préparation des données
X = data.iloc[:, 0:8]
y = data.iloc[:, 8]
```

```python
# Imprimer l'aperçu des colonnes incluses dans X
X.head()
```

```python
# Imprimer les colonnes incluses dans y
y.head()
```

## 1.Création des modèles de classification

### 1.1 Régression logistique *(Logistic Regression)*

```python
# Importer les libraires spécifiques
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
```

```python
lr_model = LogisticRegression(max_iter=1000)
lr_results = cross_val_score(lr_model, X, y, cv=10, scoring='accuracy')
print(f"Accuracy : {round(lr_results.mean()*100.0, 2)}% +/- {round(lr_results.std()*100.0, 2)}%")
```

### 1.2 Analyse linéaire discriminante *(Linear Discriminant Analysis - LDA)*

```python
# Importer les libraires spécifiques
from sklearn.model_selection import cross_val_score
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
lda_model = LinearDiscriminantAnalysis()
lda_results = cross_val_score(lda_model, X, y, cv=10, scoring='accuracy')
print(f"Accuracy : {round(lda_results.mean()*100.0, 2)}% +/- {round(lda_results.std()*100.0, 2)}%")
```

### 1.3 K plus proches voisins *(k-Nearest Neighbors - KNN)*

```python
# Importer les libraires spécifiques
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_model = KNeighborsClassifier()
knn_results = cross_val_score(knn_model, X, y, cv=10, scoring='accuracy')
print(f"Accuracy : {round(knn_results.mean()*100.0, 2)}% +/- {round(knn_results.std()*100.0, 2)}%")
```

### 1.4 Classification naïve bayesienne *(Naive Bayes Classifier)*

```python
# Importer les libraires spécifiques
from sklearn.model_selection import cross_val_score
from sklearn.naive_bayes import GaussianNB
```

```python
nb_model = GaussianNB()
nb_results = cross_val_score(nb_model, X, y, cv=10, scoring='accuracy')
print(f"Accuracy : {round(nb_results.mean()*100.0, 2)}% +/- {round(nb_results.std()*100.0, 2)}%")
```

### 1.5 Arbres de décision *(Decision Trees)*

```python
# Importer les libraires spécifiques
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
```

```python
dt_model = DecisionTreeClassifier()
dt_results = cross_val_score(dt_model, X, y, cv=10, scoring='accuracy')
print(f"Accuracy : {round(dt_results.mean()*100.0, 2)}% +/- {round(dt_results.std()*100.0, 2)}%")
```

### 1.6 Machines à vecteurs de support *(Support Vector Machine - SVM)*

```python
# Importer les libraires spécifiques
from sklearn.model_selection import cross_val_score
from sklearn.svm import SVC
```

```python
svc_model = SVC()
svc_results = cross_val_score(svc_model, X, y, cv=10, scoring='accuracy')
print(f"Accuracy : {round(svc_results.mean()*100.0, 2)}% +/- {round(svc_results.std()*100.0, 2)}%")
```

## 2. Comparer les algorithmes d'apprentissage automatique

### 2.1 Regrouper les modèles dans un tableau

C'est la méthode statique de regrouper les résultats obtenus pour les modèles. 

```python
# Créer une liste avec le nom des algorithmes
algo = ['Logistic Regression', 
        'Linear Discriminant Analysis', 
        'k-Nearest Neighbors', 
        'Naive Bayes Classifier', 
        'Decision Trees Classifier', 
        'Support Vector Machine']
```

```python
# Créer une liste avec le nom des modèles
modname = ['lr_model', 
           'lda_model', 
           'knn_model', 
           'nb_model', 
           'dt_model', 
           'svc_model']
```

```python
# Créer une liste avec les précisions
acc = []
acc.append(lr_results.mean())
acc.append(lda_results.mean())
acc.append(knn_results.mean())
acc.append(nb_results.mean())
acc.append(dt_results.mean())
acc.append(svc_results.mean())
acc
```

```python
# Créer une liste avec les écart-types
std = []
std.append(lr_results.std())
std.append(lda_results.std())
std.append(knn_results.std())
std.append(nb_results.std())
std.append(dt_results.std())
std.append(svc_results.std())
std
```

```python
# Créer un dictionnaire avec les listes qu'on vient de créer
results = {'algorithm':algo, 'model_name':modname, 'accuracy':acc, 'std':std}
print(results)
```

```python
results
```

```python
# Importer le dictionnaire dans un tableau de données pandas
results_table = pd.DataFrame(results)
results_table = results_table.round({'accuracy': 3, 'std': 3})
results_table = results_table.sort_values(by='accuracy', ascending=False)
results_table
```

### 2.2 Rouler tous les modèles en même temps avec une boucle

Une façon plus efficace d'obtenir les résultats pour tous les algorithmes en même temps est d'utiliser une boucle. 

```python
# Préparer les modèles et les enregistrer dans une liste
models = []
models.append(('LR', LogisticRegression(max_iter=1000)))
models.append(('LDA', LinearDiscriminantAnalysis()))
models.append(('KNN', KNeighborsClassifier()))
models.append(('DT', DecisionTreeClassifier()))
models.append(('NB', GaussianNB()))
models.append(('SVM', SVC()))
```

```python
models
```

```python
# Initialiser 2 listes pour y enregistrer les résultats de la boucle
results = []
names = []

# Créer la boucle pour rouler chaque modèle
for name, model in models:
  cv_results = cross_val_score(model, X, y, cv=10, scoring='accuracy')
  results.append(cv_results)
  names.append(name)
  print_out = f"{name}: {round(cv_results.mean(), 3)} {round(cv_results.std(), 3)}"
  print(print_out)
```

```python
print(results)
```

```python
print(names)
```

## 3. Visualiser les performances des modèles

### 3.1 En utilisant matplotlib

```python
# Importer les librairies de visualisation
import matplotlib.pyplot as plt
import seaborn as sns

# Définir le style des graphiques
sns.set_style("darkgrid")
```

```python
# Visualiser avec matlplotlib
fig = plt.figure()
ax = fig.add_subplot()
plt.boxplot(results)
ax.set_xticklabels(names)
plt.title('Comparaison des algorithmes', fontsize=20)
plt.xlabel('Modèles', fontsize=15)
plt.ylabel('Précision', fontsize=15)
plt.show()
```

### 3.2 En utilisant seaborn

Pour utiliser la librairie seaborn, il faut tout d'abord créer un tableau qui contient les résultats dans un format long. 

```python
# Créer un dictionnaire contenant les modèles et les résultats de la validation croisée
dict(zip(names, results))
```

```python
# Créer un tableau de données avec le dictionnaire des résultats
scores_table = pd.DataFrame(data=dict(zip(names, results)))
scores_table
```

```python
# Transformer les données du format large au format long
scores_table_melt = pd.melt(scores_table, var_name='model', value_name='score')
scores_table_melt.head(11)
```

```python
# Visualiser avec seaborn
sns.boxplot(data=scores_table_melt, x='model', y='score')
plt.title('Comparaison des algorithmes', fontsize=20)
plt.xlabel('Modèles', fontsize=15)
plt.ylabel('Précision', fontsize=15)
plt.show()
```
