#### 1.3.2 Dictionnaires (Bonus)

Les dictionnaires sont une combinaison **"clé-valeur"** (key-value) qui peuvent être modifiées en tout temps.

```python
# Définir un dictionnaire
mydict = {'a': 1, 'b': 2, 'c': 3}
mystudents = {"alex": 170, "vincent": 156}
print(mydict)
```

```python
# Imprimer la valeur associée à la clé a
print(f'La valeur associée à la clé a est: {mystudents["alex"]}')
```

```python
# Changer la valeur associée à une clé
mydict['a'] = 10
print(mydict)
print(f'La nouvelle valeur associée à a est: {mydict["a"]}')
```

```python
# Imprimer toutes les clés dans une liste
print(f'Clés: {list(mystudents.keys())}')
print(f'Clés: {mystudents.keys()}')
```

```python
# Imprimer toutes les valeurs dans une liste
print(f'Valeurs: {list(mystudents.values())}')
```

```python
# Imprimer chaque clé et valeur du dictionnaire, une à la fois
for key, value in mydict.items():
    print(f'Clé: {key}, Valeur: {value}')
```

#### 2.1 Définir un array.

Qu'est-ce qu'un array numpy? Quels sont ses avantages?
* Elles sont équivalentes aux listes Python.
* Elles accélèrent les processus et analyses.
* Elles peuvent être utilisées sur plusieurs dimensions (1D, 2D ou 3D).

*Note: On ne traitera pas des arrays 3D dans le cadre du cours.*



```python
# Définir un array simple avec 1 dimension (1D)
mylist = [1, 2, 3]
print(mylist)
myarray = np.array(mylist)
print(myarray)

# Imprimer un array simple 1D et sa forme (shape)
print(f'Contenu de l\'array 1D : {myarray}')
print(f'Forme de l\'array 1D : {myarray.shape}')
```

```python
# Définir un array matriciel avec 2 dimensions (2D)
mylist = [[1, 2, 3], [4, 5, 6]]
myarray = np.array(mylist)

# Imprimer un array simple 1D et sa forme (shape)
print(f'Contenu de l\'array 2D : \n {myarray}')
print(f'Forme de l\'array 2D : {myarray.shape}')
```

## 3. Matplotlib - Cours rapide

Objectifs de cette section:
1. Créer un graphique linéaire
2. Créer un graphique à points

```python
# Importer les librairies matplotlib et numpy
import matplotlib.pyplot as plt
import numpy as np
```

#### 3.1 Créer un graphique linéaire de base

```python
# Graphique linéaire de base
myarray = np.array([1, 2, 3])
plt.plot(myarray)
plt.xlabel('titre axe des x')
plt.ylabel('titre axe des y')
plt.title('titre du graphique')
plt.show()
```

#### 3.2 Créer un graphique à points de base

```python
# Graphique à points de base
x = np.array([1, 2, 3])
y = np.array([2, 4, 6])
plt.scatter(x, y, c="green", marker="^")
plt.title('titre du graphique')
plt.xlabel('titre axe des x')
plt.ylabel('titre axe des y')
plt.show()
```

## 4. Pandas - Cours rapide

Objectifs de cette section:
1. Comprendre les séries (structure de données utilisée par pandas)
2. Introduction aux tables de données (dataframe)

```python
# Importer les librairies numpy et pandas
import numpy as np
import pandas as pd 
```

### 4.1 Comprendre les séries avec Pandas

```python
# Utilisation des séries
myarray = np.array([16, 16, 17])
rownames = ['thomas', 'alex', 'sam']
myseries = pd.Series(myarray, index=rownames)
print(myseries)
```

### 4.2 Introduction aux tables de données

```python
# Présentation d'une table de données (dataframe)

# Créer un array numpy 2D
myarray = np.array([['Adam', 16, 165], ['Leah', 15, 147], ['Maxime', 18, 181], ['Jessica', 12, 132]])

# Définir le nom des lignes (rows)
rownames = ['row_0', 'row_1', 'row_2', 'row_3']

# Définir le nom des colonnes (columns)
colnames = ['Nom', 'Age', 'Grandeur']

# Générer la table de données
mydata = pd.DataFrame(myarray, index=rownames, columns=colnames)

# Imprimer la table de données
print(mydata)
```

### 4.3 Accéder aux données d'une table de données (dataframe)

Objectifs :
1. Obtenir les valeurs d'une colonne.
2. Obtenir les valeurs d'une ligne.
3. Obtenir une valeur précise.

Il y a plusieurs différentes méthodes pour extraire des données d'une table, soit par nom/texte ou par index/position.

Méthodes :
1. Sélectionner **une colonne** simple avec **dataframe.[colname]** ou **dataframe.colname**
2. Sélectionner **plusieurs colonnes** avec **dataframe.[[colnames]]**.
3. Opérateur **.loc[ ]** pour sélectionner une ou plusieurs **ligne** par **nom**
4. Opérateur **.loc[ ]** pour sélectionner une ou plusieurs **colonnes** par **nom**
5. Opérateur **.iloc[ ]** pour sélectionner une ou plusieurs **ligne** par **position**.
6. Opérateur **.iloc[ ]** pour sélectionner une ou plusieurs **colonnes** par **position**.

Infos sur **.loc[ ]**: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.loc.html 

Infos sur **.iloc[ ]**: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.iloc.html

#### 4.3.1 Sélectionner **une colonne** simple avec **dataframe.[colname]** ou **dataframe.colname**

Ces deux méthodes sont équivalentes et vont produire le même résultat.

```python
# Imprimer les valeurs de la colonne 2
print('Méthode 1')
print(mydata['Nom'])
```

```python
# Imprimer les valeurs de la colonne 2
print('Méthode 2')
print(mydata.Nom)
```

#### 4.3.2 Sélectionner **plusieurs colonnes** avec **dataframe.[[colnames]]**

Cette méthode permet de sélection plus d'une colonne à la fois.

```python
# Imprimer les valeurs des colonne 2 et 3
print(mydata[['Nom', 'Grandeur']])
```

#### 4.3.3 Opérateur **.loc[ ]** pour sélectionner une ou plusieurs ligne par **nom**

```python
# Imprimer les valeurs d'une seule ligne par nom
print(mydata.loc['row_2'])
```

```python
# Imprimer les valeurs de plusieurs lignes par nom
print(mydata.loc[['row_2', 'row_3']])
```

#### 4.3.4 Opérateur **.loc[ ]** pour sélectionner une ou plusieurs colonnes par **nom**

Quelques petites précisions:
1. Le `[:, ]` a l'intérieur des crochet et avant la virgule signifie que toutes les lignes seront sélectionnées.
2. Le `[,:]` a l'intérieur des crochet et avant la virgule signifie que toutes les colonnes seront sélectionnées.

```python
# Imprimer les valeurs de toutes les lignes pour UNE colonne
print(mydata.loc[:, 'Age'])
```

```python
# Imprimer les valeurs de toutes les lignes pour PLUSIEURS colonnes
print(mydata.loc[:, ['Nom', 'Age']])
```

#### 4.3.5 Opérateur **.iloc[ ]** pour sélectionner une ou plusieurs ligne par **position**

```python
# Imprimer les valeurs d'une seule ligne par position
print(mydata.iloc[1])
```

```python
# Imprimer les valeurs de plusieurs lignes par position
print(mydata.iloc[0:3])
```

#### 4.3.6 Opérateur **.iloc[ ]** pour sélectionner une ou plusieurs colonnes par **position**

```python
# Imprimer les valeurs de toutes les lignes pour UNE colonne
print(mydata.iloc[:, 1])
```

```python
print(mydata.loc[:, 'Age'])
```

```python
# Imprimer les valeurs de toutes les lignes pour PLUSIEURS colonnes
print(mydata.iloc[:, 0:2])
```

#### 4.3.7 Sélectionner un nombre précis de colonnes et de lignes

Sélectionner les colonnes Nom et Grandeur pour la première et dernière ligne.

```python
# En utilisant loc[] (nom)
print(mydata.loc[['row_0','row_3'], ['Nom','Grandeur']])
```

```python
# En utilisant iloc[] (position)
print(mydata.iloc[[0,3], [0,2]])
```
